### Problem 2b: 
Given: $$\dot{x} = -y + \mu x + xy^2$$
$$ \dot{y} = x + \mu y - x^2$$

To plot: phase portraits and trajectories for different values of $\mu$

```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# Defining the model
def model(t, z, mu):
    dxdt = -z[1] + mu*z[0] + z[0]*z[1]**2
    dydt = z[0] + mu*z[1] - z[0]**2
    
    return [dxdt, dydt]

# # Defining the Jacobian
# def mode_jac(t, z, mu):
#     dxdx = mu + z[1]**2
#     dxdy = -1 + 2*z[0]*z[1]
#     dydx = 1 - 2*z[0]
#     dydy = mu

#     return np.array([[dxdx,dxdy],[dydx,dydy]])
```

```python
# Generating intiial conditions

dataSet_length = 10
x_in = []
y_in = []

# for i in range(dataSet_length):
#     x_in.append(np.random.uniform(-0.5, 0.5))
#     y_in.append(np.random.uniform(-0.5, 0.5))
    
samples = 20

x = np.linspace(-1.5, 1.5, samples)
y = np.linspace(-1.5, 1.5, samples)
X,Y = np.meshgrid(x, y)
    
```

```python
# The following initial conditions were randomlt chosen accoring to the above cell
# The intitial condition random generation has been commented out to freeze the intiial conditions everytime the notebook is rerun

x_in = [-0.23, 0.45, 0.39, -0.1, 0.14, -0.17, -0.44, -0.33, 0.12, -0.21] 
y_in = [ 0.02, -0.05, 0.29, 0.39, -0.3, -0.37, -0.11, 0.38, 0.31, -0.43]
```

```python
# Solving the system for mu = -0.1

mu = -0.1
t_start = 0
t_end = 20
t = np.linspace(t_start, t_end, 1000)

solution = []

for i in range(len(x_in)):
    solution.append(solve_ivp(model, [t_start,t_end], [x_in[i], y_in[i]], args = (mu,), dense_output=True))

    
DX,DY = model(0, [X,Y], mu)
nrm = np.sqrt(DX**2 + DY**2)
    
fig, ax = plt.subplots()
ax.quiver(X, Y, DX/nrm, DY/nrm, np.log(nrm))

for i in range(len(solution)):
    x_sol = solution[i].sol(t).T[:, 0]
    y_sol = solution[i].sol(t).T[:, 1]
    ax.plot(x_sol, y_sol, lw="1", label = "($x_0$, $y_0$) = (%.2f, "%x_in[i] + "%.2f)"%y_in[i])
    plt.arrow(x_sol[7], y_sol[7], x_sol[7]-x_sol[6], y_sol[7] - y_sol[6], shape='full', lw=2, length_includes_head=False, head_width=.03)

ax.legend(loc = "best")
ax.grid()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
fig.suptitle("Phase portrait for $\mu$ = %.2f"%mu)
```
