## CHIRAG VERMA
### Problem sheet 2, problem 3d.

To solve: $$\dot{x}_1 = x_2$$ $$\dot{x}_2 = -\frac{k}{m}x_1 - \frac{c}{m}x_2$$

<br> and $$\dot{y}_1 = py_2$$ $$\dot{y}_2 = -y_1 - y_2$$

```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Defining the parameters c, k, m
c, k, m = 0.5, 2, 10
```

```python
# Solving the dimensionalized system

# Defining the timestamp
t_D = np.linspace(0, 300, 3000)

# Defining the initial conditions: z0_D = [x_1(0), x_2(0)]
z0_D = [5, 5] 

# The state vector is defined as z = [x_1(t), x_2(t)]

# Defining the system
def harmOsc_D(z, t):
    dx1 = z[1]
    dx2 = (-k/m)*z[0] + (-c/m)*z[1]
    return [dx1, dx2]

# Solving for the system 
z = odeint(harmOsc_D, z0_D, t_D)
```

```python
# Plotting the time variation of state variables
x1 = z[:, 0]
x2 = z[:, 1]

fig1, ax = plt.subplots()

ax.plot(t_D, x1, label = "$x_1$")
ax.plot(t_D, x2, label = "$x_2$")

ax.legend()
fig1.suptitle('Time Variation of state variables')
ax.set_xlabel('$t$')
ax.set_ylabel('$x_1$ or $x_2$') 
ax.grid()
```

```python
# Plotting the trajectory of the system
fig2, ax = plt.subplots()

plt.plot(x1, x2)

fig2.suptitle('Trajectory of the system')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$') 
ax.grid()

plt.arrow(x1[4], x2[4], x1[4]-x1[3], x2[4] - x2[3], shape='full', lw=4, length_includes_head=False, head_width=.1)

```

```python
# Defining parameters

p = 80
alpha = 2

gamma = k/(c*p)
beta = alpha/(p*gamma)
```

```python
# Solving the non-dimensionalized system

# Defining the timestamp
t_nD = np.linspace(0, 300, 3000)*gamma

# Defining the initial conditions: z0_D = [y_1(0), y_2(0)]
z0_nD = [alpha*z0_D[0], beta*z0_D[1]] 

# The state vector is defined as z = [y_1(tprime), y_2(tprime)]

# Defining the system
def harmOsc_nD(z, t):
    dy1 = p*z[1]
    dy2 = -z[0] - z[1]
    return [dy1, dy2]

# Solving for the system 
z = odeint(harmOsc_nD, z0_nD, t_nD)
```

```python
# Plotting the time variation of state variables
y1 = z[:, 0]
y2 = z[:, 1]

fig3, ax = plt.subplots()

ax.plot(t_nD, y1, label = "$y_1$")
ax.plot(t_nD, y2, label = "$y_2$")

ax.legend()
fig3.suptitle('Time Variation of state variables')
ax.set_xlabel('$t^\prime$')
ax.set_ylabel('$y_1$ or $y_2$') 
ax.grid()
```

```python
# Plotting the trajectory of the system
fig4, ax = plt.subplots()

# Multiplying the final non dimensionalized solutions by respective constants to scale up the graph
plt.plot(y1/alpha, y2/beta, linestyle=(3, (5, 5)), label = "non-Dimensionalized")
plt.plot(x1, x2, linestyle=(1, (5, 5)), label = "Dimensionalized")

fig4.suptitle('Trajectory of the system')
ax.set_xlabel('$y_1$ or $x_1$')
ax.set_ylabel('$y_2$ or $x_2$') 
ax.grid()
ax.legend(loc = "best")

# plt.arrow(y1[4], y2[4], y1[4]-y1[3], y2[4] - y2[3], shape='full', lw=4, length_includes_head=False, head_width=.1)

```
