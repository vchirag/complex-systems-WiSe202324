### Prob 4d
To simulate: the trajectories for $p = 0.999$ and $p = 1.101$

```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Plotting for p = 1.101
# Defining the system
def system(theta, t):
    return np.sin(theta) + 1.001

# Defining the timestamp
t_ = np.linspace(0, 500, 1000)

# Defining the intiial condition
theta_in = list(np.arange(0, 2*np.pi, np.pi/5))

# State vector which stores value of theta in time for specific values of initial conditions
theta_state = []

# Solving the system
for i in range(len(theta_in)):
    theta_state.append(odeint(system, theta_in[i], t_))

fig, ax  = plt.subplots()

for i in range(len(theta_in)):
    plt.plot(t_, theta_state[i], label = r"$\theta_{in}$ = %.3f"%theta_in[i])

    
plt.xlim([0, t_[-1]])
plt.ylim(0, 25)

ax.legend(loc = "best")
ax.axhline(y = 0.0, linewidth = 2.5, color = 'k')
ax.axvline(x = 0.0, linewidth = 2.5, color = 'k')
ax.set_xlabel("$t$")
ax.set_ylabel(r"$\theta$")
fig.suptitle(r"Evolution of $\theta(t)$ for p = 1.001")
ax.grid()
```
