### Problem 2 b:
Given: $$\dot n  = rn\left(1 - \frac{n}{k}\right) - \frac{n^2}{1+n^2}$$

To plot: the intersections between two function on the right hand side of the equation for $n \neq 0$



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the first function

def func1(r, n, k = 10):
    return r*(1 - n/k)

def func2(n):
    return n/(1 + n**2)
```

```python
# Steps for the horizontal axis

# Since a population can not be negative, the stamps start from zero
n_ = np.arange(0, 15, 0.1)

# Defining various values of r
r_ = [-0.1, 0.1, 0.37,  0.5, 0.57, 1]

# Storing the value of the func1 in a list for different values of r
op_func1 = []

for i in range(len(r_)):
        op_func1.append(func1(r_[i], n_))


# Storing the value of function 2
op_func2 = func2(n_)
```

```python
fig, ax = plt.subplots()

for i in range(len(r_)):
    plt.plot(n_, op_func1[i], label = "r = %.2f"%r_[i], linewidth = '2')

plt.plot(n_, op_func2, linewidth = 3, color = "k", label=r"$\frac{n}{1+n^2}$")
plt.xlim([0, n_[-1]])
plt.ylim([-0.25, 1.25])


ax.axhline(y = 0.0, linewidth = 1, color = 'k')
ax.axvline(x = 0.0, linewidth = 1, color = 'k')
    
ax.grid()

ax.legend(loc = "best")
ax.set_xlabel("$n$")
# ax.set_ylabel("$(x, r)$")
```
