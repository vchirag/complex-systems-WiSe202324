TO Graphically discuss hte imperfect supercritical bifurcation for 
$$\dot{x} = px - x^3 + h$$

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x, p, h):
    return p*x - x**3 +h
```

```python
x_ = np.arange(-4, 4, 0.1)
p_ = [-1, 0, 1, 2, 5]
h_ = [-1, 0, 2/3*np.sqrt(3), 2, 4*np.sqrt(2)/3*np.sqrt(3), 5]
```

```python
fig1, ax1 = plt.subplots()

for i in range(len(h_)):
    plt.plot(x_, func(x_, p_[0], h_[i]), label = "p = %.2f, "%p_[0] + "h = %.2f"%h_[i])

ax1.axhline(y = 0.0, linewidth = 1, color = 'k')
ax1.axvline(x = 0.0, linewidth = 1, color = 'k')
    
ax1.grid()

ax1.legend(loc = "best")
ax1.set_xlabel("$n$")
fig1.suptitle("$\dot{x} = px - x^3 + h$ for p=%.2f and varying h"%p_[0])
```
