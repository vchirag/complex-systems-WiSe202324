### Prob 4 a

Given: $\theta \in [0, 2\pi]$, and
$$\dot {\theta} = f(\theta) = p + \sin(\theta)$$

To plot: $f(\theta)$ for different values of p

```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the function

def func(theta, p):
    return p + np.sin(theta)
```

```python
theta_ = np.arange(0, 2*np.pi, np.pi/100)
p_ = list(np.arange(-1.5, 2, 0.5))

op_func = []

for i in range(len(p_)):
    op_func.append(func(theta_, p_[i]))
```

```python
fig, ax  = plt.subplots()

for i in range(len(p_)):
    plt.plot(theta_, op_func[i], label = "p = %.1f"%p_[i])
    
plt.xlim([0, theta_[-1]])
plt.ylim(p_[0] - 1.5, p_[-1] + 1.5)

ax.legend(loc = "best")
ax.axhline(y = 0.0, linewidth = 2.5, color = 'k')
ax.axvline(x = 0.0, linewidth = 2.5, color = 'k')
    
ax.grid()
```
