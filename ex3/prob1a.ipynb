### Problem 1 a:

Given: $$\dot{x} = x(r - e^x)\hspace{1cm}$$
       and 
       $$\dot{y} = y^4 - 4y + \xi$$
       where $r$ and $\xi$ are paramters
       
To Plot: The Right Hand Side around and at the bifurcation

```python
# importing libraries

import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Defining functions

# Defining the first function, viz. the function with parameter r
def func_x(x, r):
    return x*(r - np.exp(x))

# Defining the second function, viz. the function with parameter xi
def func_y(y, xi):
    return y**4 - 4*y + xi
```

```python
# Defining the stamps for the horizontal axis
x_ = np.arange(-5, 5, 0.1)

# A list to store the value of the function for correponding values of paramters
# op_x = [[@r = -0.5, func @ r = -0.5], [@r = 0, func @ r = 0], [@r = 0.1, func @ r = 0.1], [@r = 0.1, func @ r = 0.1], [@r = 0.1, func @ r = 0.1]]
op_x = []

r_ = [-0.5, 0, 0.1, 1, 5]

for i in range(len(r_)):
    op_x.append([r_[i], func_x(x_, r_[i])])
```

```python
# Plotting the function with the paramter r

fig1, ax = plt.subplots()

for i in range(len(r_)):
    plt.plot(x_, op_x[i][1], label = "r = %.1f"%op_x[i][0])

ax.axhline(y = 0.0, linewidth = 1.5, color = 'k')
ax.axvline(x = 0.0, linewidth = 1.5, color = 'k')
    
ax.set_ylim(-5, 5)
ax.grid()

ax.legend(loc = "best")
ax.set_xlabel("$x$")
ax.set_ylabel("$F_1(x, r)$")
fig1.suptitle("Bifurcation @ $r = 1$")
```

```python
# Defining the stamps for the horizontal axis
y_ = np. arange(-2, 2.5, 0.1)

# A list to store the value of the function for correponding values of paramters
# op_y = [[@xi = -2, func @ xi = -2], [@xi = 0, func @ xi = 0] [@xi = 1, func @ xi = 1], [@xi = 3, func @ xi = 3], [@xi = 5, func @ xi = 5]]
op_y = []
xi_ = [-2, 0, 1, 3, 5]

for i in range(len(xi_)):
    op_y.append([xi_[i], func_y(y_, xi_[i])])
```

```python
# Plotting the function with the paramter xi

fig2, ax = plt.subplots()

for i in range(len(xi_)):
    plt.plot(y_, op_y[i][1], label = r"$\xi$ = %.2f"%op_y[i][0])

ax.axhline(y = 0.0, linewidth = 1.5, color = 'k')
ax.axvline(x = 0.0, linewidth = 1.5, color = 'k')
    
ax.set_ylim(-5, 5)
ax.grid()

ax.legend(loc = "best")
ax.set_xlabel("$y$")
ax.set_ylabel(r"$F_2(y, \xi)$")
fig2.suptitle(r"Bifurcation @ $\xi = 3$")
```
